```python
import pandas as pd
import numpy as np
```

**WITHIN THREE YEARS -> VEHAGE**

```python
df = pd.read_csv('Dataset.csv')
df.loc[df['VEHAGE'] <= 0, 'VEHAGE'] = np.nan
df = df.dropna()
df = df[df['VEHAGE'] <= 3]
df
```

**HFUEL (Class Target)**

```python
df.loc[df['HFUEL'] < -1, 'HFUEL'] = np.nan
df=df.dropna()
conditions = [(df['HFUEL'] == 1),(df['HFUEL'] >= 2) & (df['HFUEL'] <= 4),(df['HFUEL'] > 4),(df['HFUEL'] == -1)]
values = [0,1, 0, 0]
df['HFUEL'] = np.select(conditions, values)
df
```

**HHFAMINC**

```python
df.loc[df['HHFAMINC'] < 0, 'HHFAMINC'] = np.nan
df=df.dropna()
conditions = [(df['HHFAMINC'] <= 3),(df['HHFAMINC'] >= 4) & (df['HHFAMINC'] <= 5),(df['HHFAMINC'] == 6),(df['HHFAMINC'] == 7),(df['HHFAMINC'] >= 8)]
values = [1,2,3,4,5]
df['HHFAMINC'] = np.select(conditions, values)
df
```
